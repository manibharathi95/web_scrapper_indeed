```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys

# https://selenium-python.readthedocs.io/waits.html
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

import pandas as pd
import os

path = "C:\Program Files (x86)\chromedriver"

driver = webdriver.Chrome(path)
driver.get("https://in.indeed.com/?from=gnav-jobsearch--jasx")


print(driver.title)

what = driver.find_element_by_id("text-input-what")
what.send_keys("data scientist")

where = driver.find_element_by_id("text-input-where")
where.send_keys("India")
where.send_keys(Keys.RETURN)

titles=[]
companies=[]
locations=[]
links =[]
reviews=[]
salaries = []

for i in range(1,5):
    

    # xpath syntax: //tagname[@attributeName='value'] (or) //tagname[contains(@attributeName, 'value')]

    job_card = driver.find_elements_by_xpath('//div[contains(@class,"clickcard")]')

    # print(job_card)

    for job in job_card:
       
    #  REVIEWS
        try:
            review = job.find_element_by_xpath('.//span[@class="ratingsContent"]').text
        except:
            review = "None"
        reviews.append(review)
   
    #   SALARY
        try:
            salary = job.find_element_by_xpath('.//span[@class="salaryText"]').text
        except:
            salary = "None"    
        salaries.append(salary)

    #   JOB LOCATION    
        try:
            location = job.find_element_by_xpath('.//span[contains(@class,"location")]').text
        except:
            location = "None"    
        locations.append(location)
        
    #   JOB TITLE        
        try:
            title  = job.find_element_by_xpath('.//h2[@class="title"]//a').text
        except:
            title = job.find_element_by_xpath('.//h2[@class="title"]//a').get_attribute(name="title")
        titles.append(title)
        
    #   APPLICATION LINK
        links.append(job.find_element_by_xpath('.//h2[@class="title"]//a').get_attribute(name="href"))
        
    #   COMPANY NAME    
        companies.append(job.find_element_by_xpath('.//span[@class="company"]').text)
        
    
    try:
        next_page = driver.find_element_by_xpath('//a[@aria-label={}]//span[@class="pn"]'.format(i+1))
        next_page.click()

    except:
        next_page = driver.find_element_by_xpath('//a[@aria-label="Next"]//span[@class="np"]')
        next_page.click()
         
    print("Page: {}".format(str(i+1)))

driver.quit()

pwd = os.getcwd()
#print(pwd)
master_dict = {"Job Title": titles, "Name of the Ompany": companies, "Location": locations, "Salary": salaries, "Review": reviews, "Link": links}
df = pd.DataFrame(master_dict)
print(df)
df.to_excel(pwd + "\Final_Output.xlsx", index=False)
```
